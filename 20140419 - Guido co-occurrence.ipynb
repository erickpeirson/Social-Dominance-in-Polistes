## Generate f/N for each 'concept' over time

```python
# { concept: time-period: f }
termcounts = { ex[0]:Counter() for ex in explananda+explanantes }
```

```python
# Pooled f for all concepts.
em_counts = Counter()
```

```python
# We will also generate co-occurrence graphs.
results = GraphCollection()
```

```python
for period,dois in D.axes['date'].iteritems():
    results.graphs[period] = nx.Graph()
    pairs = {}    # Keep track of node pairs.
    nodes = Counter()    # f (concept) for this time-period; used to generate 
                         #  'count' attribute in graph.
    
    for doi in dois:
        these_ex = []    # Explananda in this document.
        these_es = []    # Explanantes in this document.
        em_found = False
        
        # Look for explananda.
        for explanandum in explananda:
            ex_found = False
            for ex_variant in explanandum:
                ex_type = len(ex_variant.split())    # 1, 2, or 3.
                
                try:    
                    if ex_variant in types[ex_type][doi] and types[ex_type][doi][ex_variant] > 1:
                        # Count explanandum (concept) only once/document.
                        if not ex_found:
                            nodes[explanandum[0]] += 1  
                            these_ex.append(explanandum[0])    
                            termcounts[explanandum[0]][period] += 1 
                            ex_found = True
                        
                        # Count presence of explananda (all) only once/document.
                        if not em_found:
                            em_counts[period] += 1
                            em_found = True
                except KeyError:
                    pass    # Occassionally a DOI is missing.
        
        # Look for explanantes.
        for explanans in explanantes:
            es_found = False
            for es_variant in explanans:
                es_type = len(es_variant.split())    # 1, 2, or 3.
                try:                
                    if es_variant in types[es_type][doi] and types[es_type][doi][es_variant] > 1:
                        # Count explananda (concept) only once/document.
                        if not es_found:
                            nodes[explanans[0]] += 1  
                            these_es.append(explanans[0])    
                            termcounts[explanans[0]][period] += 1       
                            es_found = True
                except KeyError:
                    pass    # Occassionally a DOI is missing.

        # Generate co-occurrence graph data for this document.
        for ex in these_ex:
            for es in these_es:
                e_f = (ex, es)
                e_b = (es, ex)
                if e_f in pairs.keys():
                    pairs[e_f] += 1
                elif e_b in pairs.keys():
                    pairs[e_b] += 1
                else:
                    pairs[e_f] = 1                

    # Build co-occurrence graph for each time-period.
    for pair,count in pairs.iteritems():
        for n,v in nodes.iteritems():
            results.graphs[period].add_node(n, count=v)
        results.graphs[period].add_edge(pair[0], pair[1], weight=count)
        nx.set_node_attributes(results.graphs[period], 'type', {k:v for k,v in e_index.iteritems() if k in nodes.keys()})
```

## Frequency of each explanans in the sub-corpus containing explananda: $p(es\big| EM)  = {f_{es} \over N_{EM}}, \forall es\in ES$

```python
N_EM = {}
N_em = Counter()
```

```python
for period,dois in D.axes['date'].iteritems():
    N_EM[period] = { em[0]:set([]) for em in explananda }
    for doi in dois:
        em_found = False
        for em in explananda:
            e = em[0]
            e_type = len(e.split())
            try:
                if e in types[e_type][doi] and types[e_type][doi][e] > 1:
                    N_EM[period][e].add(doi)
                    if not em_found:
                        N_em[period] += 1
                        em_found = True
            except KeyError:
                pass
```

## Frequency of each explanans in the sub-corpus containing each explandum:
$p(es\big| em) = {f_{es} \over N_{em}}$

### and Normalized Pointwise Mutual Information:
$nPMI(es,em) = { ln{p(es,em) \over p(es)p(em)} \over -ln p(es,em) }$

```python
T = range(1872, 2011)
```

```python
f_matrix = np.zeros((N_s, N_m, len(T)))
N_matrix = np.zeros((N_s, N_m, len(T)))
nPMI = np.zeros((N_s, N_m, len(T)))
```

```python
f_es = { es[0]:Counter() for es in explanantes }
for i_p in xrange(len(range(1872,2011))):
    period = range(1872,2011)[i_p]
    EM = N_EM[period]
    for i_s in xrange(N_s):
        s = explanantes[i_s][0]
        s_type = len(s.split())
        for i_m in xrange(N_m):
            m = explananda[i_m][0]
            dois = EM[m]    # Get only documents that contain this explanandum.
            for doi in dois:
                if s in types[s_type][doi] and types[s_type][doi][s] > 1:    # Contains this explanans?
                    f_matrix[i_s, i_m, i_p] += 1.    
            N_matrix[i_s, i_m, i_p] = len(dois)
```

```python
# Generates some NaNs where N_matrix is 0.
fN_matrix = np.nan_to_num(f_matrix/N_matrix)
```

```python
# Calculate nPMI
for i_p in xrange(len(T)):
    period = range(1872,2011)[i_p]
    for i_s in xrange(N_s):
        es = explanantes[i_s][0]
        for i_m in xrange(N_m):
            em = explananda[i_m][0]
            p_x = tf[es][period]
            p_y = tf[em][period]
            
            # Normalize for entire corpus.
            p_xy = f_matrix[i_s, i_m, i_p]/len(D.axes['date'][period])

            if p_x != 0 and p_y != 0:
                pmi = (log( p_xy /( p_x * p_y ) ))/( -1 * log(p_xy) )                                
                if pmi > 0.:
                    nPMI[i_s, i_m, i_p] = pmi
```

```python
fig = plt.figure(figsize=(30,20))
for i_s in xrange(N_s):
    for i_m in xrange(N_m):

        sp = i_m + 1 + i_s*N_m
        ax = plt.subplot(N_s, N_m, sp)
        ax.plot(T, fN_matrix[i_s, i_m, :], label=r'$p(es| em) = \frac{f_{es}}{N_{em}}$', lw=3, alpha=0.6)
        ax.plot(T, nPMI[i_s, i_m, :], 'g--', label=r'$nPMI(es,em)$', alpha=0.8)        
        ax.set_ylim(0., 1.0)
        ax.set_xlim(1872, 2011)

        if i_m == 0:
            ax.set_ylabel(explanantes[i_s][0], fontsize=16)
            
        if i_s == 0:
            ax.set_title(explananda[i_m][0], fontsize=16)
            
        if i_s == 0 and i_m == N_m - 1:
            ax.legend(bbox_to_anchor=(1.6,1))
plt.show()
```
