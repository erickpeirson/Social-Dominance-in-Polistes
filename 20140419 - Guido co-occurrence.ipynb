```python
import tethne.readers as rd
import tethne.writers as wr
import tethne.analyze as az
from tethne.data import DataCollection, GraphCollection
import networkx as nx
import os
from collections import Counter
import matplotlib.pyplot as plt
```

```python
unigrams = {}
bigrams = {}
trigrams = {}
datapath = "/Users/erickpeirson/Dropbox/Digital Hps/Files/data"
for directory in os.listdir(datapath):
    unigrams.update(rd.dfr.ngrams(datapath + "/" + directory, 'uni'))
    bigrams.update(rd.dfr.ngrams(datapath + "/" + directory, 'bi'))
    trigrams.update(rd.dfr.ngrams(datapath + "/" + directory, 'tri'))    
```

```python
papers = rd.dfr.from_dir(datapath)
```

```python
explananda = [  ('polygyny',), 
                ('division of labor',), 
                ('hierarchy','hierarchies','hierarchical'),                               
                ('social dominance',),
                ('dominance',)]
```

```python
explanantes = [ ('ovaries', 'ovary'), 
                ('relatedness',), 
                ('hormones', 'hormone', 'hormonal'), 
                ('endocrine',), 
                ('fitness',), 
                ('cost','costs',), 
                ('benefit', 'benefits'),
                ('ontogeny', 'ontogenesis', 'ontogenetic', 'developmental' ) ]
```

```python
explananda
```

```python
e_index = {}
for ex in explananda:
    e_index[ex[0]] = 'explananda'    
for es in explanantes:
    e_index[es[0]] = 'explanantes'
```

```python
e_index
```

```python
print len(papers), len(unigrams), len(bigrams), len(trigrams)
```

```python
D = DataCollection(papers, index_by='doi')
```

```python
D.slice('date', method='time_window', window_size=4, step_size=1)
```

```python
types = {1: {d:{k:v for k,v in vals} for d,vals in unigrams.iteritems()},
         2: {d:{k:v for k,v in vals} for d,vals in bigrams.iteritems()},
         3: {d:{k:v for k,v in vals} for d,vals in trigrams.iteritems()}}
```

```python
results = GraphCollection()
```

```python
plt.figure(figsize=(10,5))
plt.hist([len(v) for v in unigrams.values()], bins=60, label='unigrams')
plt.hist([len(v) for v in bigrams.values()], bins=60, label='bigrams')
plt.hist([len(v) for v in trigrams.values()], bins=60, label='trigrams')
plt.legend()
plt.show()
```

```python
termcounts = { ex[0]:Counter() for ex in explananda+explanantes }
```

```python
for period,dois in D.axes['date'].iteritems():
    results.graphs[period] = nx.Graph()
    pairs = {}
    nodes = Counter()
    for doi in dois:
        these_ex = []
        these_es = []
        for explanandum in explananda:
            for ex_variant in explanandum:
                ex_type = len(ex_variant.split())
                try:
                    if ex_variant in types[ex_type][doi]:
                        nodes[explanandum[0]] += 1  
                        these_ex.append(explanandum[0])    
                        termcounts[explanandum[0]][period] += 1 
                except:
                    pass
        for explanans in explanantes:
            for es_variant in explanans:
                es_type = len(es_variant.split())
                try:                
                    if es_variant in types[es_type][doi]:
                        nodes[explanans[0]] += 1  
                        these_es.append(explanans[0])    
                        termcounts[explanans[0]][period] += 1       
                except:
                    pass                    
        for ex in these_ex:
            for es in these_es:
                e_f = (ex, es)
                e_b = (es, ex)
                if e_f in pairs.keys():
                    pairs[e_f] += 1
                elif e_b in pairs.keys():
                    pairs[e_b] += 1
                else:
                    pairs[e_f] = 1                

    for pair,count in pairs.iteritems():
        for n,v in nodes.iteritems():
            results.graphs[period].add_node(n, count=v)
        results.graphs[period].add_edge(pair[0], pair[1], weight=count)
        nx.set_node_attributes(results.graphs[period], 'type', {k:v for k,v in e_index.iteritems() if k in nodes.keys()})
```
